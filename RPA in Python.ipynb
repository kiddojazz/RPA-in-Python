- Automatically Download Data from a site

The first thing we need to do is to import the rpa package and initialize the rpa object.

```python
r.init()
```

This would automatically open your Chrome browser and navigate to the UCI website to download the breast cancer dataset.

```python
r.url("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/")
```

Note: when you get to a site the next step is for you to click on the breast-cancer.data

```python
r.click("breast-cancer.data")
```

Note that if there are two hyperlinks with the same label “here” and we execute rpa.click("breast-cancer.data"),
then the robot will click on the first one it finds.

```python
r.close()
```

- Web Automation

```python
r.init()
```

```python
r.url('https://www.google.com')
r.type('//*[@name="q"]', 'USA[enter]')
print(r.read('result-stats'))
r.snap('page', 'USA.png')
```

```python
r.close()
```
